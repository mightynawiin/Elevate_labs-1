```python
import pandas as pd
from datetime import datetime

# Loading the  dataset
df = pd.read_csv('/content/marketing_campaign.csv', delimiter='\t')

# Cleaning the column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace(r'\W', '', regex=True)

# Convert dt_customer to datetime and format as dd-mm-yyyy
df['dt_customer'] = pd.to_datetime(df['dt_customer'], dayfirst=True, errors='coerce')
df['dt_customer'] = df['dt_customer'].dt.strftime('%d-%m-%Y')

# Calculating the age
current_year = datetime.now().year
df['age'] = current_year - df['year_birth']

# Checking for missing values
print("Missing values per column:")
print(df.isnull().sum())

# Droping the duplicates if any
df = df.drop_duplicates()

#Strip whitespace from object/string columns
for col in df.select_dtypes(include='object'):
    df[col] = df[col].str.strip()

print(df)
print(df.isnull().sum())
mean_income = df['income'].mean()
df['income'] = df['income'].fillna(mean_income)
print(df['income'].isnull().sum())
# Save cleaned data
df.to_csv('cleaned_dataset.csv', index=False)
```
